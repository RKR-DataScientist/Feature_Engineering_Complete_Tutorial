# One Hot Encoding & Dummy Variables | Categorical Variable Encoding

Machine Learning algorithm cant work on categorical data so we have to encode categorical variables in a numerical format.

But we can do one-hot encoding on only "Nominal Categrical features". 

**A nominal category or a nominal group is a group of objects or ideas that can be collectively grouped on the basis of a particular characteristic—a qualitative property.**

Nominal data are used to label variables without any quantitative value. Common examples include male/female (albeit somewhat outdated), hair color, nationalities, names of people, and so on.

## Using pandas to do OneHot Encoding

```
# Import libraries
import pandas as pd
```

```
df = pd.read_csv(r'/content/raw.csv')
df
```

```
# Let's create dummies varible
dummy_df = pd.get_dummies(df)
dummy_df
```

Now you can earlier we had 7 columns, but now after dummies varible, it get converted as 0 and 1 form with 13 columns.

But point is that it called **K-Varible**, means we have 1 and 0 form, but we can drop 0 or 1, any one of them, becuase we can understand that 1 would male and 0 would female, so we can drop it.

**Therefore, to get out from Dummy Varible Trap** We would use K-1 ,means drop_first=True

pd.get_dummies(data, prefix=None, prefix_sep='_', dummy_na=False, columns=None, sparse=False, drop_first=False, dtype=None) -> 'DataFrame'

```
pd.get_dummies(df, drop_first=True)
```

Now dummy trap varible  has been removed

## OneHot Encoding using Scikit-Learn

Syntex

OneHotEncoder(categories='auto', drop=None, sparse=True, dtype=<class 'numpy.float64'>, handle_unknown='error')

```
from sklearn.preprocessing import OneHotEncoder
```

```
# Let's create object for onehot encoding
on_hot = OneHotEncoder(sparse=False)

# Now we need to tranform the data
on_enc_arr = on_hot.fit_transform(df[['sex',	'smoker',	'day',	'time']])

# we will get data into numpy array format
on_enc_arr
```

```
# Now let's create dataframe and put all into columns
on_enc_df = pd.DataFrame(on_enc_arr, columns=)
on_enc_df
```

```
# In this dataframe, we couldn't get columns, we need to assign manually
dummy_df.keys()
```

```
on_enc_df = pd.DataFrame(on_enc_arr, columns=['sex_Female', 'sex_Male', 'smoker_No',
       'smoker_Yes', 'day_Fri', 'day_Sat', 'day_Sun', 'day_Thur',
       'time_Dinner', 'time_Lunch'])
on_enc_df
```
